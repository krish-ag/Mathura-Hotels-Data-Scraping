```python
import pandas as pd
import json
```

```python
with open('data.json') as f:
    data = json.load(f)
```

```python
hotels = data.get('response').get('personalizedSections')[0].get('hotels') + data.get('response').get('personalizedSections')[1].get('hotels')
```

```python
hotels[0]
```

```python
# json.dump(hotels, open('hotels.json', 'w'))
```

```python
hotelDetails = []

for hotel in hotels:
    hotelDetails.append({
        'name': hotel.get('name'),
        'propertyType': hotel.get('propertyType'),
        'starRating': hotel.get('starRating'),
        'locationPersuasionText': hotel.get('locationPersuasion'),
        'priceWithoutTax': hotel.get('priceDetail').get('price') if 'priceDetail' in hotel else None,
        'discountedPrice': hotel.get('priceDetail').get('discountedPrice') if 'priceDetail' in hotel else None,
        'discountedPriceWithTax': hotel.get('priceDetail').get('discountedPriceWithTax') if 'priceDetail' in hotel else None,
        'totalTax': hotel.get('priceDetail').get('totalTax') if 'priceDetail' in hotel else None,
        'totalReviewCount': hotel.get('reviewSummary').get('totalReviewCount') if 'reviewSummary' in hotel else None,
        'totalRatingCount': hotel.get('reviewSummary').get('totalRatingCount') if 'reviewSummary' in hotel else None,
        'locationDetails': hotel.get('locationDetail').get('name') if 'locationDetail' in hotel else None,
        'geoLocation': hotel.get('geoLocation'),
        'Link': hotel.get('detailDeeplinkUrl'),
        'poiTag': hotel.get('poiTag'),
    })
```

```python
df = pd.DataFrame(hotelDetails)
```

```python
# df null values
df[df.isnull().any(axis=1)]
```

```python
df.to_json('hotelDetails.json', orient='records')
```
